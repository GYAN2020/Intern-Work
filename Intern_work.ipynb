```python
#1.Drop unwanted columns
df=df.drop(columns=['index','PendingS'],errors='ignore')
```

```python
#convert Date column to datetime
df['Date']=pd.to_datetime(df['Date'],format='%d-%m-%Y',errors='coerce')
```

```python
#3. Check for missing values
print("Missing Values:")
print(df.isnull().sum())
```

```python
#Preview after cleaning
df.head()
```

```python
#1.Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
#2. remove rows with missing Amount or Currency
df=df.dropna(subset=['Amount','currency'])
```

```python
df.shape
```

```python
#drop unnecessary column
df=df.drop(columns=['New'],errors='ignore')
```

```python
#4. drop rows missing shipping data
df=df.dropna(subset=['ship-city','ship-state','ship-postal-code','ship-country'])
```

```python
#check again
print("Data after cleaning:")
print(df.isnull().sum())

print("\n Remaining Rows:",len(df))
```

```python
#total revenue
total_revenue=df['Amount'].sum()
print(f"Total Revenue:{total_revenue:,.2f}")
```

```python
#Average order value
avg_order_value=df['Amount'].mean()
print(f"Average order value:{avg_order_value:,.2f}")
```

```python
#Total quantity sold
total_qty=df['Qty'].sum()
print(f"Total Quantity Sold:{total_qty:,}")
```

```python
#Orders by Status
order_status=df['Status'].value_counts()
print("\n Order Status distribution:")
print(order_status)
```

```python
#Montly sales trend
df['Month']=df['Date'].dt.to_period('M')
monthly_sales=df.groupby('Month')['Amount'].sum()
```

```python
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='line',marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.savefig("monthly_sales_trend.jpg")
plt.show()

```

```python
#customer segmentation by city (proxy)
city_segment=df.groupby('ship-city').agg({
    'Order ID':'nunique',#total unique orders
    'Amount':['sum','mean'],#total and average spend
    'Qty':'sum'            #total items bought
})
```

```python
#flatten multi-level columns
city_segment.columns=['Total_Orders','Total_Sales','Avg_Spend','Total_Quantity']
city_segment=city_segment.sort_values(by='Total_Sales',ascending=False).head(15)
```

```python
#visualization -Sales vs Avg Spend
plt.figure(figsize=(10,6))
plt.scatter(city_segment['Total_Sales'],city_segment['Avg_Spend'],s=city_segment['Total_Orders']/10)
for city in city_segment.index:
    plt.text(city_segment.loc[city,'Total_Sales'],city_segment.loc[city,'Avg_Spend'],city,fontsize=9)
plt.title("City-wise segmentation: Total sales vs Avg Spend per order")
plt.xlabel("Total Sales")
plt.ylabel("Average Spend")
plt.grid(False)
plt.savefig("City_wise_segmentation.png", dpi=300)
plt.show()
```

```python




# Clean theme
plt.style.use('seaborn-v0_8')
sns.set_palette("Blues_r")

# Load your dataset
df = pd.read_csv("Amazon Sale Report.csv")

# Clean missing and convert types
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.drop_duplicates()
df = df.dropna(subset=['Amount', 'currency'])
df = df.dropna(subset=['ship-city','ship-state','ship-postal-code','ship-country'])

# Create time-based columns
df['Month'] = df['Date'].dt.to_period('M')
df['Year'] = df['Date'].dt.year
df['Month_num'] = df['Date'].dt.month



#Category Distribution
plt.figure(figsize=(12,5))
df['Category'].value_counts().head(15).plot(kind='bar')
plt.title("Category Distribution (Top 15)")
plt.xlabel("Category")
plt.ylabel("Order Count")
plt.xticks(rotation=45)
plt.savefig("Category_distribution.png", dpi=300)
plt.show()


# Price vs Quantity Scatter Plot 
plt.figure(figsize=(10,6))
plt.scatter(df['Qty'], df['Amount'], alpha=0.6)
plt.title("Price vs Quantity Scatter Plot")
plt.xlabel("Quantity")
plt.ylabel("Amount (₹)")
plt.grid(True)
plt.show()


# Size Preference Charts
size_counts = df['Size'].value_counts()

plt.figure(figsize=(10,5))
size_counts.plot(kind='bar')
plt.title("Size Preference Distribution")
plt.xlabel("Size")
plt.ylabel("Order Count")
plt.xticks(rotation=0)
plt.savefig("Size_preference.png", dpi=300)
plt.show()


size_sales = df.groupby('Size')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
size_sales.plot(kind='bar', color='green')
plt.title("Sales Amount by Size")
plt.ylabel("Revenue (₹)")
plt.xticks(rotation=0)
plt.show()

```
